```python
import sqlite3
import pandas as pd
import yfinance as yf
import datetime
con = sqlite3.connect('mydatabase.db')
cursor = con.cursor()
cursor.execute('DROP TABLE fifteen_MIN')
cursor.execute('CREATE TABLE fifteen_MIN(Ticker, Datetime, Open, High, Low, Close, Adj_Close, Volume)')
```

```python
for symbol in sp500.Symbol:
    start = datetime.datetime(2000, 1, 1)
    end = datetime.datetime(2022, 11, 27)
    df = yf.download(symbol, start=start, end=end, group_by='ticker')

    df['Ticker'] = symbol
    #print(df)

    df.columns = df.columns.str.replace('Adj Close', 'Adj_Close')
    df.to_sql('fifteen_MIN', con, if_exists='append', index_label='Datetime')
```

```python
import random
tickers = pd.read_sql('SELECT DISTINCT Ticker FROM Price', con)

ticker = np.random.choice(tickers['Ticker'].tolist())

```

```python
dates = pd.read_sql('SELECT Datetime FROM Price WHERE Ticker="'+ticker+'"', con)
dates = pd.to_datetime(dates['Datetime'])
start_date = dates.max()
end_date = dates.min()

#
time_between_dates = start_date-end_date

days_between_dates = int(time_between_dates.days-365*2.5)

random_number_of_days = random.randrange(days_between_dates)
random_date = end_date + datetime.timedelta(days=random_number_of_days)
#
print(random_date)
```
